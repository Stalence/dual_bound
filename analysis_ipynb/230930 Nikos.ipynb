```python
rows = []

for problem, hugedict in x.items():
    k_solution_sizes = []
    k_times = []
    l_solution_sizes = []
    l_times = []
    for dataname, information in hugedict.items():
        _ , kamis, lwd = information
        assert kamis[0] == 'kamis'
        assert lwd[0] == 'lwd'
        
        N = dataname.split("_")[1]
        maybeN = dataname.split("_")[2]
        if N != maybeN: print(dataname)
        if int(N) != 500: print(dataname)

        k_solution_sizes.append(int(N) - kamis[1])
        k_times.append(kamis[2])
        l_solution_sizes.append(int(N) - lwd[1])
        l_times.append(lwd[2])
    row = dict(zip(['problem', "k_solution_sizes","l_solution_sizes", "k_times", "l_times"], [problem, k_solution_sizes, l_solution_sizes, k_times, l_times] ))
    rows.append(row)
```

```python
df = pd.DataFrame.from_records(rows)
```

```python
import numpy as np
dfkeylist = df.keys()
savekeys = []
for k in dfkeylist:
    if k == 'problem':continue
    df[f'{k}_average'] = df[k].apply(np.mean)
    savekeys.append(f'{k}_average')
```

```python
df[['problem'] + savekeys].to_csv('kamis_lwd_baselines.csv')
```
